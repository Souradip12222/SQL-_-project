```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Q1

```python
df = pd.read_csv('new_wheels_project_question_1.csv')
```

```python
df.head()
```

```python
df.head(5).plot.bar(x='STATE', y='CUSTOMER_COUNT', rot=0)
plt.title("Top 5 States with Most Customers")
plt.xlabel("State")
plt.ylabel("Customers")
plt.xticks(rotation=30, horizontalalignment="center");
```

# Q2

```python
df1.head()
```

# Q3

```python
df2 = pd.read_csv('new_wheels_project_question_3.csv')
```

```python
df2.head()
```

```python
df2.plot(kind='bar', stacked=True, x='QUARTER_NUMBER', rot=0, colormap='Set2')
plt.title("Trend of Customer Satisfaction")
plt.xlabel("Quarter Number")
plt.ylabel("Average Customer Rating (%)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0);
```
